##Exploratory Analysis and Visualization

###Country

```python
country_names=df2.Country.value_counts().index
country_names
```

```python
Country_value=df2.Country.value_counts().values
Country_value
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#Top 3 Country that uses Zomato shown in Pie Chart
plt.pie(Country_value[:3],labels=country_names[:3],autopct='%1.1f%%', radius=2.2, 
        textprops={'fontsize': 20}, )
```

Observation:

 Zomato maximum record is  from india
2nd United States and than
3rd United Kingdom 

###Ratings

```python
ratings=df2.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Obeservation

1. Rating from 4.5 to 4.9 --> Excellent
2. Rating from 4.0 to 3.4 --> very good
3. Rating from 3.5 to 3.9 --> good
4. Rating from 3.0 to 3.4 --> average
5. Rating from 2.0 to 2.4 --> poor
6. Rating of 0.0 --> not rated

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(15,6)
sns.barplot(x="Aggregate rating", y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating", y="Rating Count",data=ratings,hue='Rating color',palette=['white','red','orange','yellow','green', 'darkgreen'])
```

Observation

1. Maximum number of ratings are from 2.5 to 3.4
2. No rated count is High

```python
#count plot
sns.countplot(x="Rating color",data=ratings,palette=['white','red','orange','yellow','green', 'darkgreen'])
```

```python
df2.groupby(['Aggregate rating','Country']).size().reset_index()
```

Observation

1. Maximum number of 0 rating are from indian coustomer
